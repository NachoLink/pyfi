# Downloading a file from the internet - US Census Retail Sales

```python
import pandas as pd
```

```python
#If the data you need has a download link, right click and "Copy link address..."
#In this case, we are downloading Retail Sales Data from the US Census
#It is a good idea to download the file first to understand its structure and contents: https://www.census.gov/retail/mrts/www/mrtssales92-present.xls

#Define the url
url = "https://www.census.gov/retail/mrts/www/mrtssales92-present.xls"

#Read the excel file and pull the 
retail_sales = pd.read_excel(url, sheet_name='2017')

retail_sales.head(10)
```

```python
#Download the file, define sheet_name and header row, reset index, restrict to rows 1-68, skip first column, rename Sales_Type column
retail_sales = pd.read_excel(url, sheet_name='2017', header=4).reset_index().iloc[1:68,1:].rename(columns = {'level_1': 'Sales_Type'})

#Melt the data frame on Sales_Type as ID, rename variable/value columns, drop rows with >= 2 NAs
retail_sales = retail_sales.melt(id_vars=['Sales_Type'], var_name='Month', value_name='Sales_$MM').dropna(thresh=2)

retail_sales.head(10)

```
